```python
# create TensorFlow session
sess = gpt2.start_tf_sess()

# load the model
gpt2.load_gpt2(sess, run_name=name_of_finetuned_model)
```

```python
# create Prefix for QA model
ctx = input("Context: ")
```

```python
qst = input("Question: ")
ans = "[ANSWER]:"
pre = '<|startoftext|>\n[CONTEXT]: ' + ctx + "\n[QUESTION]: " + qst + "\n" + ans
```

```python
# create Prefix for math model
equ = input("Equation: ")
solveFor = input("solve for: ")

pre = '<|startoftext|>\n[EQUATION]: ' + equ + "\n[SOLVEFOR]: " + solveFor
```

```python
# generate output
gpt2.generate(sess, run_name=name_of_finetuned_model, prefix=pre, temperature=0.2, length=20, nsamples=100, batch_size=10, truncate="<|endoftext|>")
```
